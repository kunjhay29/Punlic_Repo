Importing the necessary libraries for webscraping and retrieving info from the web

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

writing a function to make request and get the data in the html tags 

necessary error handling procedure utilized.

```python
def table_head(url):
    r =  requests.get(url)
    global table
    global df
    if r.status_code == 200:
        try:
            st = datetime.now()
            Soup = BeautifulSoup(r.text, "html")
            table = Soup.find('table')
            titles = table.find('tr')
            table_titles = [name.text for name in titles]
            del table_titles[0]
            df = pd.DataFrame(columns = table_titles)
            print(datetime.now() - st)
            return df
        except:
            pass
    elif r.status_code == 429:
        print('Too many request!!!!, kindly take a break and try again')
    elif r.status_code == 500:
        print("Server Error!!!")
    else:
        print("No reponse kindly confirm root")
```

Function to join tanle headers and body of the table inside a dataframe

```python
def new_table():
    Column_data = table.find_all('tr')
    for row in Column_data[1:]:
        row_data = row.find_all('td')
        Individual_row = [data.text for data in row_data[1:]]
        length = len(df)
        df.loc[length] = Individual_row
    return df
```

calling the function to get table headers

```python
table_head('https://www.investing.com/indices/major-indices')
```

calling the function to get the full content of the scraped table data

```python
new_table()
```
